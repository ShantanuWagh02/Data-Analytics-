### Total Sales by Region and Category
- Create a table/output showing the sum of sales for each Region(Rows level) and Category(Columns).
- Fill missing values with 0 and include row/column totals.

```python
sales_region_cat = pd.pivot_table(data=df1,
               index='Region',
               columns= 'Category',
               values='Sales',
               aggfunc='sum',  # aggregate function
               fill_value=0 ,    # null value fill with 0
               margins=True, #add total value 'All'   column
               margins_name='grand total' ) # change column name 'All' to 'Grand total'

sales_region_cat
```

### Sales and Profit Summary by Category
- Build a pivot that shows, for each Category, both total and average of Sales and Profit.
- Display Category as index

```python
df1.pivot_table(
               index='Category',
               values=['Sales','Profit'],
               aggfunc=['sum','mean'],
               margins=True,
               margins_name='Grand Total')
```
